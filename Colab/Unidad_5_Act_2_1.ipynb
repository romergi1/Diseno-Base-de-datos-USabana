## Data Preprocessing

### Convert Timestamp Columns to Datetime

```python
datetime_cols_orders = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in datetime_cols_orders:
    df_orders[col] = pd.to_datetime(df_orders[col])
display(df_orders.info())
```

```python
df_order_items['shipping_limit_date'] = pd.to_datetime(df_order_items['shipping_limit_date'])
display(df_order_items.info())
```

### Handle Missing Values

```python
# For df_orders, fill missing `order_approved_at` with `order_purchase_timestamp`
df_orders['order_approved_at'] = df_orders['order_approved_at'].fillna(df_orders['order_purchase_timestamp'])

# For `order_delivered_carrier_date` and `order_delivered_customer_date`, fill missing values with `order_estimated_delivery_date`.
df_orders['order_delivered_carrier_date'] = df_orders['order_delivered_carrier_date'].fillna(df_orders['order_estimated_delivery_date'])
df_orders['order_delivered_customer_date'] = df_orders['order_delivered_customer_date'].fillna(df_orders['order_estimated_delivery_date'])

display(df_orders.info())
```

```python
# For df_products
# Fill missing product_category_name with 'unknown'
df_products['product_category_name'] = df_products['product_category_name'].fillna('unknown')

# Fill numerical columns with their median
for col in ['product_name_lenght', 'product_description_lenght', 'product_photos_qty',
            'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']:
    df_products[col] = df_products[col].fillna(df_products[col].median())

display(df_products.info())
```

```python
# For df_order_reviews, fill missing comment titles and messages with empty strings
df_order_reviews['review_comment_title'] = df_order_reviews['review_comment_title'].fillna('')
df_order_reviews['review_comment_message'] = df_order_reviews['review_comment_message'].fillna('')

display(df_order_reviews.info())
```

### Further Data Type Conversion and Missing Value Handling in Merged DataFrame

```python
# Convert review timestamp columns to datetime
datetime_cols_reviews = ['review_creation_date', 'review_answer_timestamp']
for col in datetime_cols_reviews:
    df_merged[col] = pd.to_datetime(df_merged[col], errors='coerce') # Use 'coerce' to turn unparseable dates into NaT

# Fill missing product_category_name_english with 'unknown_category_english'
df_merged['product_category_name_english'] = df_merged['product_category_name_english'].fillna('unknown_category_english')

# For remaining numerical columns introduced by left merges (e.g., from order_items, payments, reviews),
# fill NaN values with 0 or an appropriate indicator, depending on their meaning.
# For example, order_item_id, price, freight_value, payment_sequential, payment_installments, payment_value, review_score.
# If an order has no items, the item-specific columns will be NaN. Filling with 0 makes sense if it implies 'no value'.
# Similarly for payments and reviews, if an order has no corresponding entry, it implies no payment/review data.

# Numerical columns to fill with 0 (assuming NaN means absence/zero value)
numerical_cols_to_fill = [
    'order_item_id', 'price', 'freight_value',
    'product_name_lenght', 'product_description_lenght', 'product_photos_qty',
    'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm',
    'seller_zip_code_prefix', 'payment_sequential', 'payment_installments', 'payment_value',
    'review_score'
]

for col in numerical_cols_to_fill:
    df_merged[col] = df_merged[col].fillna(0)

# Fill remaining object type columns with 'unknown' or empty string where appropriate
object_cols_to_fill = [
    'product_id', 'seller_id', 'seller_city', 'seller_state',
    'payment_type', 'review_id', 'review_comment_title', 'review_comment_message'
]

for col in object_cols_to_fill:
    if col in df_merged.columns:
        df_merged[col] = df_merged[col].fillna('unknown')

# Handle remaining specific missing values after merges and initial filling
# Fill missing 'shipping_limit_date' with 'order_purchase_timestamp'
df_merged['shipping_limit_date'] = df_merged['shipping_limit_date'].fillna(df_merged['order_purchase_timestamp'])

# Fill missing 'product_category_name' with 'unknown'
df_merged['product_category_name'] = df_merged['product_category_name'].fillna('unknown')

# Fill remaining NaT values in review timestamp columns with 'order_purchase_timestamp'
df_merged['review_creation_date'] = df_merged['review_creation_date'].fillna(df_merged['order_purchase_timestamp'])
df_merged['review_answer_timestamp'] = df_merged['review_answer_timestamp'].fillna(df_merged['order_purchase_timestamp'])

display(df_merged.info())
```

### Define Product Attributes

```python
# 1. Merge df_products with df_product_category_name_translation
# This gives us product_id, product_category_name, and product_category_name_english
df_product_catalog_temp = pd.merge(df_products, df_product_category_name_translation,
                                   on='product_category_name', how='left')

# Fill any remaining missing product_category_name_english (e.g., if 'unknown' category wasn't translated)
df_product_catalog_temp['product_category_name_english'] = df_product_catalog_temp['product_category_name_english'].fillna('unknown_category_english')

# 2. Calculate average price per product_id from df_order_items
# This ensures a single price entry per product if a product appears in multiple order items with slightly varying prices
df_product_prices = df_order_items.groupby('product_id')['price'].mean().reset_index()
df_product_prices.rename(columns={'price': 'avg_price'}, inplace=True) # Rename for clarity before final merge

# 3. Merge the product details with the calculated average prices
df_final_products = pd.merge(df_product_catalog_temp, df_product_prices,
                             on='product_id', how='left')

# 4. Select and rename columns to match the requested attributes
# Use 'product_category_name_english' for both 'name' (as a proxy) and 'category'
df_products_attributes = df_final_products[[
    'product_id',
    'product_category_name_english',
    'avg_price'
]].copy()

df_products_attributes.rename(columns={
    'product_id': '_id',
    'product_category_name_english': 'category',
    'avg_price': 'price'
}, inplace=True)

# Add 'name' column using 'category' as a proxy
df_products_attributes['name'] = df_products_attributes['category']

# Reorder columns to desired order: _id, name, category, price
df_products_attributes = df_products_attributes[['_id', 'name', 'category', 'price']]

# 5. Handle missing 'price' values for products that never appeared in an order item
df_products_attributes['price'] = df_products_attributes['price'].fillna(0)

display(df_products_attributes.head())
display(df_products_attributes.info())
```

### Exploring Data with MongoDB Queries and Aggregation Pipelines

We will now use `pymongo` to perform various queries and aggregation pipelines on our `products` and `reviews` collections. These examples demonstrate how to retrieve specific data, filter, sort, and calculate aggregated metrics, leveraging the structured and embedded document design we've implemented.

#### Query 1: Find Products by Category, Sorted by Price

This query demonstrates how to find all products belonging to a specific category and then sort them by their price in ascending order. We'll project only the relevant fields (`_id`, `name`, `category`, `price`).

```python
# Get the 'products' collection
products_collection = db['products']

# Define the category to search for
category_to_search = 'health_beauty'

# Perform the query
print(f"Products in category '{category_to_search}', sorted by price (top 5):\n")
for product in products_collection.find(
    {'category': category_to_search},
    {'_id': 1, 'name': 1, 'category': 1, 'price': 1}
).sort('price', 1).limit(5):
    display(product)
```

#### Query 2: Find Products with High Average Ratings (e.g., >= 4.5)

This query uses the pre-calculated `metrics.average_rating` field to find products that have received high average scores, sorted in descending order of rating.

```python
print("Products with an average rating of 4.5 or higher (top 5):\n")
for product in products_collection.find(
    {'metrics.average_rating': {'$gte': 4.5}},
    {'_id': 1, 'name': 1, 'category': 1, 'metrics.average_rating': 1, 'price': 1}
).sort('metrics.average_rating', -1).limit(5):
    display(product)
```

#### Aggregation Pipeline 1: Top 5 Products by Total Units Sold

This pipeline retrieves the top 5 products based on their `metrics.total_units_sold`. It demonstrates accessing nested fields in aggregation.

```python
print("Top 5 products by total units sold:\n")
pipeline_top_sold = [
    {'$sort': {'metrics.total_units_sold': -1}},
    {'$limit': 5},
    {'$project': {
        '_id': 1,
        'name': 1,
        'category': 1,
        'total_units_sold': '$metrics.total_units_sold'
    }}
]

for doc in products_collection.aggregate(pipeline_top_sold):
    display(doc)
```

#### Aggregation Pipeline 2: Average Review Score per Product Category

This pipeline calculates the average review score for each product category. It groups documents by `category`, computes the average `metrics.average_rating`, and then sorts the results.

```python
print("Average review score per product category (top 5 categories):\n")
pipeline_avg_rating_category = [
    {'$match': {'category': {'$ne': 'unknown_category_english'}}},
    {'$group': {
        '_id': '$category',
        'average_review_score': {'$avg': '$metrics.average_rating'}
    }},
    {'$sort': {'average_review_score': -1}},
    {'$limit': 5}
]

for doc in products_collection.aggregate(pipeline_avg_rating_category):
    display(doc)
```

### Design Flexible 'specifications' Subdocument

```python
# 1. Select the relevant product specification columns from df_products
specification_cols = [
    'product_id',
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]
df_product_specs = df_products[specification_cols].copy()

# 2. Convert each row of specifications into a dictionary, grouped by product_id
def create_specs_dict(row):
    # Exclude 'product_id' from the specifications dictionary
    return {col: row[col] for col in specification_cols if col != 'product_id'}

df_product_specs['specifications'] = df_product_specs.apply(create_specs_dict, axis=1)

# 3. Create a DataFrame with just product_id and the new 'specifications' column
df_product_specs_final = df_product_specs[['product_id', 'specifications']].copy()
df_product_specs_final.rename(columns={'product_id': '_id'}, inplace=True)

# 4. Merge this back with df_products_attributes
# This will add the 'specifications' subdocument to our product catalog
df_products_with_specs = pd.merge(df_products_attributes, df_product_specs_final,
                                  on='_id', how='left')

display(df_products_with_specs.head())
display(df_products_with_specs.info())
```

### Update Product Collection in MongoDB with Specifications

```python
# Get the 'products' collection
products_collection = db['products']

# Iterate through the DataFrame and update each document in MongoDB
updated_count = 0
for index, row in df_products_with_specs.iterrows():
    product_id = row['_id']
    specifications = row['specifications']

    # Update the document in MongoDB
    result = products_collection.update_one(
        {'_id': product_id},
        {'$set': {'specifications': specifications}}
    )
    if result.modified_count > 0:
        updated_count += 1

print(f"Updated {updated_count} product documents with specifications in the 'products' collection.")

# Verify a sample document to check the update
print("\nSample document after update:")
# Find one document that has specifications and display it
sample_product = products_collection.find_one({'specifications': {'$exists': True}})
if sample_product:
    display(sample_product)
else:
    print("No product found with specifications.")
```
